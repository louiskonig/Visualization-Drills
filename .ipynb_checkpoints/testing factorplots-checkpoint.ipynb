```python
def plot_hline(y,**kwargs):
    data = kwargs.pop("data") #get the data frame from the kwargs
    plt.axhline(y=y, c='red',linestyle='dashed',zorder=2) #zorder places the line underneath the other points
```

```python
sns.set_style("darkgrid")

medians = tips.groupby(['time','sex','smoker'])['total_bill'].median()
nobs =  tips.groupby(['time','sex','smoker']).apply(lambda x: 'n: {}'.format(len(x)))

g = sns.factorplot(x="sex", y="total_bill",hue="smoker", col="time",data=tips, kind="box",size=8, aspect=.8, legend_out = False)
g.map_dataframe(plot_hline,y=30)
g.add_legend(label_order = g._legend_data.keys())

g.fig.subplots_adjust(top=.93)
g.fig.suptitle('Tips') # can also get the figure from plt.gcf()

#express as %:  plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()]) 

num_labels = len(g._legend_data.keys())
num_sex = tips.sex.nunique()

for ax in plt.gcf().axes:
    ax_time = ax.get_title().partition(' = ')[-1]

    for tick, label in enumerate(ax.get_xticklabels()):
        ax_sex = label.get_text()

        for j, ax_smoker in enumerate(ax.get_legend_handles_labels()[1]):
            x_offset = (j - (num_labels-1)/2) * (num_labels/((num_labels*num_sex)+num_sex-1))
            med_val = medians[ax_time, ax_sex, ax_smoker]
            num = nobs[ax_time, ax_sex, ax_smoker]

            ax.text(tick + x_offset, med_val + 0.20, num,
                    horizontalalignment='center', size='small', color='w', weight='semibold')

g.savefig('tips.png')
plt.show()
```
